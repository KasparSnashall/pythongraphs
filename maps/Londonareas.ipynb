```python
from bs4 import BeautifulSoup

svg = open('London.svg','r')
# Load the SVG map this can be found on wikimedia commons
soup = BeautifulSoup(svg,'xml')
paths = soup.findAll('path')
#find all the paths
path_style = 'font-size:12px;fill-rule:nonzero;stroke:#000000;stroke-opacity:1;stroke-width:0.1;stroke-miterlimit:4;stroke-dasharray:none;stroke-linecap:butt;marker-start:none;stroke-linejoin:bevel;fill:'
#make a blank slate
regions = []
for i in range(33):
    print paths[i]['id']
    
#make a list of the regions to make sure they are london areas and not lakes and such
```

Through a bit of trial and error its the first 33 regions which are important I do believe.

```python
for i in range(33):
    regions.append(paths[i]['id'])
print regions
```

```python
colours = ['#d73027','#fc8d59','#fee090','#e0f3f8','#91bfdb','#4575b4']
#pretty colours
```

http://colorbrewer2.org/ lots more can be found here depending on your map prefrences.

```python
#make some random colours appear you can make a very eaasy system for whatever data you may have
for p in paths:
    if p['id'] in regions:
        color_class = np.random.randint(0,5)
        color = colours[color_class]
        p['style'] = path_style + color
```

```python
svgStr = soup.prettify()
london = open('NewLondon.svg','w') 
london.write(svgStr)
london.close()
```
